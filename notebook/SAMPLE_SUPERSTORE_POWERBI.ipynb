##### 6. Cek Outlier Data

```python
# Menghitung total seluruh baris data sebagai pembagi persentase
total_data = df.shape[0]

print("=== DETAIL BATAS & PERSENTASE OUTLIER ===\n")

# ==========================================
# 1. ANALISIS OUTLIER KOLOM: SALES
# ==========================================
Q1_sales = df['Sales'].quantile(0.25)
Q3_sales = df['Sales'].quantile(0.75)
IQR_sales = Q3_sales - Q1_sales
lower_sales = Q1_sales - (1.5 * IQR_sales)
upper_sales = Q3_sales + (1.5 * IQR_sales)

# Menghitung jumlah baris yang di luar batas
outlier_sales = df[(df['Sales'] < lower_sales) | (df['Sales'] > upper_sales)].shape[0]
# Rumus persentase: (jumlah outlier / total data) * 100
pct_sales = (outlier_sales / total_data) * 100

print(f"--- Kolom Sales ---")
print(f"Batas Bawah: {lower_sales:.2f} | Batas Atas: {upper_sales:.2f}")
print(f"Jumlah Outlier: {outlier_sales} baris")
print(f"Persentase Outlier: {pct_sales:.2f}%\n")


# ==========================================
# 2. ANALISIS OUTLIER KOLOM: QUANTITY
# ==========================================
Q1_qty = df['Quantity'].quantile(0.25)
Q3_qty = df['Quantity'].quantile(0.75)
IQR_qty = Q3_qty - Q1_qty
lower_qty = Q1_qty - (1.5 * IQR_qty)
upper_qty = Q3_qty + (1.5 * IQR_qty)

outlier_qty = df[(df['Quantity'] < lower_qty) | (df['Quantity'] > upper_qty)].shape[0]
pct_qty = (outlier_qty / total_data) * 100

print(f"--- Kolom Quantity ---")
print(f"Batas Bawah: {lower_qty:.2f} | Batas Atas: {upper_qty:.2f}")
print(f"Jumlah Outlier: {outlier_qty} baris")
print(f"Persentase Outlier: {pct_qty:.2f}%\n")


# ==========================================
# 3. ANALISIS OUTLIER KOLOM: DISCOUNT
# ==========================================
Q1_disc = df['Discount'].quantile(0.25)
Q3_disc = df['Discount'].quantile(0.75)
IQR_disc = Q3_disc - Q1_disc
lower_disc = Q1_disc - (1.5 * IQR_disc)
upper_disc = Q3_disc + (1.5 * IQR_disc)

outlier_disc = df[(df['Discount'] < lower_disc) | (df['Discount'] > upper_disc)].shape[0]
pct_disc = (outlier_disc / total_data) * 100

print(f"--- Kolom Discount ---")
print(f"Batas Bawah: {lower_disc:.2f} | Batas Atas: {upper_disc:.2f}")
print(f"Jumlah Outlier: {outlier_disc} baris")
print(f"Persentase Outlier: {pct_disc:.2f}%\n")


# ==========================================
# 4. ANALISIS OUTLIER KOLOM: PROFIT
# ==========================================
Q1_profit = df['Profit'].quantile(0.25)
Q3_profit = df['Profit'].quantile(0.75)
IQR_profit = Q3_profit - Q1_profit
lower_profit = Q1_profit - (1.5 * IQR_profit)
upper_profit = Q3_profit + (1.5 * IQR_profit)

outlier_profit = df[(df['Profit'] < lower_profit) | (df['Profit'] > upper_profit)].shape[0]
pct_profit = (outlier_profit / total_data) * 100

print(f"--- Kolom Profit ---")
print(f"Batas Bawah: {lower_profit:.2f} | Batas Atas: {upper_profit:.2f}")
print(f"Jumlah Outlier: {outlier_profit} baris")
print(f"Persentase Outlier: {pct_profit:.2f}%\n")
```

## Data Manipulation

### 1. Membuat kolom baru melalui feature engineering 

##### a. Menambahkan kolom profit ratio

```python
df['Profit Ratio'] = df['Profit'] / df['Sales']
```

##### b. Menambahkan kolom profit status

```python
def profit_status(profit):
    if profit > 0:
        return 'Profit'
    elif profit == 0:
        return 'BEP'
    else:
        return 'Loss'

df['Profit Status'] = df['Profit'].apply(profit_status)
```

##### c. Menambahkan kolom Discount Category

```python
def discount_category(x):
    if x == 0:
        return 'No Discount'
    elif x <= 0.2:
        return 'Low Discount'
    elif x <= 0.4:
        return 'Medium Discount'
    else:
        return 'High Discount'

df['Discount Category'] = df['Discount'].apply(discount_category)
```

```python
print(df['Discount Category'].value_counts())
```

##### d. Menambahkan kolom tahun

```python
df['Year'] = df['Order Date'].dt.year
```

##### e. Menambahkan kolom bulan

```python
df['Month'] = df['Order Date'].dt.month_name()
```

INTERPRETASI :
- Untuk analisis selanjutnya diperlukan penambahan kolom untuk digunakan dalam analisis insight nantinya
- Untuk penambahan kolom yaitu kolom profit ratio, profit status, discount category, year, dan month
